```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import ipywidgets as widgets
```

```python
# 修改路径
cubelets_dir = '/home/jclou/loujc_work/sofia/GAMA/1300-1350/Dec-0011_09_05_arcdrift/Dec-0011_09_05_arcdrift_cubelets'

# 递归查找所有 spec.txt
spec_files = sorted(glob.glob(os.path.join(cubelets_dir, '**/*_spec.txt'), recursive=True))
print(f"找到 {len(spec_files)} 个源。")
```

```python
def plot_spec(index):
    spec_path = spec_files[index]
    source_name = os.path.basename(spec_path).replace('_spec.txt', '')
    
    # 读取 SoFiA spec 数据
    try:
        df = pd.read_csv(spec_path, sep='\s+', comment='#', header=None)
        freq_mhz = df[1] / 1e6
        flux_jy = df[2]
        
        plt.figure(figsize=(10, 5))
        plt.step(freq_mhz, flux_jy, where='mid', color='teal', linewidth=1.5)
        plt.fill_between(freq_mhz, flux_jy, step="mid", alpha=0.15, color='teal')
        
        plt.axhline(0, color='black', linestyle='--', alpha=0.3)
        plt.title(f"Source ID: {index+1} | Name: {source_name}", fontsize=14)
        plt.xlabel("Frequency (MHz)", fontsize=12)
        plt.ylabel("Integrated Flux (Jy)", fontsize=12)
        plt.grid(True, which='both', linestyle=':', alpha=0.5)
        
        # 让 X 轴显示完整的频率，不要使用科学计数法
        plt.ticklabel_format(useOffset=False, style='plain', axis='x')
        plt.show()
        
    except Exception as e:
        print(f"读取源 {source_name} 出错: {e}")

# 创建交互滑动条
interact(plot_spec, index=IntSlider(min=0, max=len(spec_files)-1, step=1, description='Source Index:'));
```
